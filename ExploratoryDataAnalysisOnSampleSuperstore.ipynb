## Exploratory Data Analysis

```python
sns.pairplot(data_cleaned)
```

#### Visualising the categories

```python
plt.figure(figsize =(6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(data_cleaned['Category'].value_counts(),labels = data_cleaned['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
data_cleaned.hist(bins =50,figsize = (20,15))
```

```python
data_cleaned['State'].value_counts()
```

```python
plt.figure(figsize = (14,14))
sns.countplot(x=data_cleaned['State'])
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(10,16))
data_cleaned.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
# computing the top categories in terms of sales from first 100 observations
top_category_a = data_cleaned.groupby('Category').Sales.sum().nlargest(n=100)

# computing the top categories in terms of profit fron first 100 companies
top_category_p = data_cleaned.groupby('Category').Profit.sum().nlargest(100)
```

```python
# plotting to see it visually
plt.style.use('seaborn')
top_category_a.plot(kind='bar',figsize=(10,5),fontsize = 14)
top_category_p.plot(kind='bar',figsize = (10,5),fontsize =14,color='red')
plt.xlabel('Category',fontsize=15)
plt.ylabel('Total Sales/Profit',fontsize = 15)
plt.title("Top category sales vs profits",fontsize=15)
plt.show()
```

```python
# computing top sub-categories in terms of sales from first 100 observations
top_subcategory_s = data_cleaned.groupby("Sub-Category").Sales.sum().nlargest(n=100)
# computing top sub-categories in terms of profit from first 100 observations
top_subcategory_p = data_cleaned.groupby("Sub-Category").Profit.sum().nlargest(n=100)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize=(10,5),fontsize=14)
top_subcategory_p.plot(kind = 'bar',figsize=(10,5),fontsize=14,color='red')
plt.xlabel('Sub-Category',fontsize=15)
plt.ylabel('Total Sales/Profits',fontsize=15)
plt.title("Top sub-category Sales vs Profit",fontsize= 15)
plt.show()
```

```python
# A MORE DETAILED VIEW
plt.figure(figsize=(14,12))
statewise = data_cleaned.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh()  # horizontally aligned
```

### The graph indicates that the Copiers and  Phones  have the highest sales as well as profit whereas tables has negative profit or loss.


#### VIsualizing the discount

```python
plt.figure(figsize=(8,7))
sns.lineplot(data_cleaned['Discount'],data_cleaned['Profit'],data = data_cleaned)
```

#### Visualizing the sales vs profit in different regions

```python
plt.figure(figsize=(6,6))
plt.title('Region')
plt.pie(data_cleaned['Region'].value_counts(),labels=data_cleaned['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

#### This graph shows that West and East have nearly same profit even though the sales in the East are less as compared to West.

### Visualizing the sales vs profit in different states

```python
# computing the top states in terms of sales from first 10 observations

top_states_s = data_cleaned.groupby("State").Sales.sum().nlargest(n=10)

# computing the top states in terms of profit from first 10 observations

top_states_p = data_cleaned.groupby("State").Profit.sum().nlargest(n=10)

plt.style.use('seaborn')
top_states_s.plot(kind='bar',figsize=(10,5),fontsize = 14)
top_states_p.plot(kind='bar',figsize=(10,5),fontsize=14,color ='red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total Sales',fontsize =15)
plt.title("Top 10 states - Sales v/s Profit")
plt.show()
```

### Checking the interdependency of Sales,Profits and Discounts

```python
plt.style.use('seaborn')
data_cleaned.plot(kind = 'scatter',figsize =(10,5),x="Sales",y="Profit",c="Discount",s=20,fontsize=16,colormap='viridis')
plt.ylabel('Total Profits',fontsize =16)
plt.title("Interdependency of Sales,Profit and Discounts",fontsize = 16)
plt.show()
```

### The graph clearly shows that if we increase discount,the profit decreases.
